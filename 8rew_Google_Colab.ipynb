# Opis

**Poniżej znajduje się robocza wersja progemu 8rew, sytemu zarządzania procesami piwowarskimi.**


## Uproszczony Opis Produkcji Piwa


Piwo powstaje z 4 składników: Woda, Słód, Chmiel i drożdże.

Opis procesów produkcji piwa różni się delikatnie z zależności od źródła, w warunkach domowych po ówcześnie przygotowaniu składników przyjmuje się, że jest ich 8:
- Zacieranie (Mashing):
  - Mieszanie z wodą: Mielony słód jest mieszany z gorącą wodą w celu przekształcenia skrobi w cukry fermentacyjne. Proces ten odbywa się w tzw. zacierniku.
  - Temperatura i czas: Temperatura i czas zacierania mają wpływ na charakterystykę piwa. Typowe wartości to 65-70°C przez około 60 minut.
- Filtracja:
  - Oddzielanie brzeczki: Po zacieraniu mieszanina (zacier) jest filtrującym procesem oddzielana od cieczy (brzeczki).
  - Wysładzanie (Lautering) zacieru, wypłukuje się pozostałe cukry ze słodu.
- Gotowanie brzeczki:
  - Gotowanie: Brzeczka jest gotowana przez około 60-90 minut.
  - Dodawanie chmielu: Chmiel jest dodawany w różnych momentach gotowania w celu uzyskania odpowiedniego smaku, aromatu i gorzkości.
- Chłodzenie brzeczki:
  - Obniżanie temperatury: Po gotowaniu brzeczka musi zostać szybko schłodzona do temperatury fermentacji, aby uniknąć kontaminacji i zapewnić odpowiednie warunki dla drożdży.
- Fermentacja:
  - Dodanie drożdży: Schłodzoną brzeczkę przelewa się do fermentora i dodaje drożdże.
  - Fermentacja cicha: Piwo fermentuje w fermentatorze przez kilka dni.
  - Fermentacja cicha: Od piwa oddziela się gęstwe drożdżową i pozostawia się na kilka do kilkunastu dni, w zależności od stylu i drożdży.
- Butelkowanie lub kegowanie:
  - Przygotowanie do rozlewu: Piwo jest delikatnie przemieszane, aby rozprowadzić ewentualne osady, a następnie butelkowane lub przelewane do kegów.
  - Dodawanie cukru: Jeśli piwo jest butelkowane, dodaje się małą ilość cukru, aby zainicjować fermentację w butelce i wytworzyć naturalne bąbelki.
- Leżakowanie: Piwo w butelkach lub kegach jest przechowywane przez kilka tygodni, aby smaki się ustabilizowały, a bąbelki się uformowały.
- Degustacja: Po odpowiednim okresie leżakowania piwo jest gotowe do degustacji i konsumpcji


## Instrukcje Użytkowania



1. Tworzenie Receptury
- Nazwa: Wprowadź nazwę swojej receptury piwa.
- Styl Piwa: Wybierz styl piwa z rozwijanej listy (uzupełniana na podstawie tabeli Styles).
- Objętość: Wprowadź objętość piwa do warzenia (w litrach).
- Współczynnik woda/słód: Podaj ile części wody ma przypadać na jedną część słodu podczas zacierania (najczęściej 3)
- Temperatura zacierania: Podaj temperaturę w jakiej zamierzasz przeprowadzać zacieranie.
- Zapisz: Kliknij, aby zapisać recepturę w bazie danych. Komunikat potwierdzi pomyślne zapisanie.
- Wyczyść: Kliknij, aby wyczyścić wszystkie pola wejściowe.
2. Zacieranie
- Wybierz Recepturę: Wybierz recepturę z rozwijanej listy (uzupełniana na podstawie tabeli Recipes).
- Wydajność Zacierania: Wprowadź szacowaną wydajność zacierania jako procent (najczęsciej 80).
- Dodawanie słodów, możesz dodać wiele:
  - Wybierz Słód: Wybierz słód z rozwijanej listy (uzupełniana na podstawie tabeli Malts).
  - Waga: Wprowadź wagę wybranego słodu (w kilogramach).
  - Dodaj Słód: Kliknij, aby dodać wybrany słód do receptury.
- Oblicz Ekstrakcję: Kliknij, aby obliczyć całkowitą ekstrakcję twojej warki. Obliczona ekstrakcja zostanie wyświetlona w stopnaich Blg.
- Zapisz: Kliknij, aby zapisać dane o słodach i obliczoną ekstrakcję w bazie danych. Istniejące wartości ekstrakcji dla receptury zostaną zaktualizowane.
- Wyczyść: Kliknij, aby wyczyścić wszystkie pola i zresetować dane.
3. Chmiel
- Wybierz Recepturę: Wybierz recepturę z rozwijanej listy (uzupełniana na podstawie tabeli Recipes).
- Dodawanie chmieli, możesz dodać wiele:
  - Wybierz Chmiel: Wybierz chmiel z rozwijanej listy (uzupełniana na podstawie tabeli Hops).
  - Waga Chmielu: Wprowadź wagę chmielu (w gramach).
  - Czas: Wprowadź czas gotowania (w minutach).
  - Dodaj Chmiel: Kliknij, aby dodać wybrany chmiel do receptury.
- Oblicz IBU: Kliknij, aby obliczyć IBU na podstawie dodanych chmieli i ich parametrów. Obliczone IBU zostanie wyświetlone.
- Zapisz: Kliknij, aby zapisać dane o chmielu i obliczone IBU w bazie danych.
- Wyczyść: Kliknij, aby wyczyścić wszystkie pola i zresetować dane dotyczące chmielu.
4. Uwarz Piwo
- Wybierz Recepturę: Wybierz recepturę z rozwijanej listy (uzupełniana na podstawie tabeli Recipes).
- Uwarz to: Kliknij, aby wyświetlić pełne informacje o wybranej recepturze, oraz dokładną instrukcje przygotowania warki.
5. Oblicz procent alkoholu
- Po procesie fermentacji, zmierz poziom BLG w piwie i wprowadź go, klikając w oblicz ABV, otrzymasz informacje o procencie alkoholu etylowego w twojej warce.

# Baza danych

## 1. Utworzenie bazy danych przy pierwszym uruchomieniu programu

```python
import sqlite3

# Create a persistent database connection
conn = sqlite3.connect('beer_recipes.db', check_same_thread=False)
cursor = conn.cursor()

# Create tables
cursor.execute('''
CREATE TABLE IF NOT EXISTS Recipes (
    ID INTEGER PRIMARY KEY,
    Name TEXT,
    Date DATE,
    Volume REAL,
    Extraction REAL,
    IBU REAL,
    MashingTemperature REAL,
    MaltWaterFactor REAL,
    ExtractionAfterFermentation REAL,
    ABV REAL
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Malts (
    ID INTEGER PRIMARY KEY,
    Name TEXT,
    Extraction REAL
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Hops (
    ID INTEGER PRIMARY KEY,
    Name TEXT,
    AlphaAcids REAL
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Styles (
    ID INTEGER PRIMARY KEY,
    Name TEXT
)
''')


cursor.execute('''
CREATE TABLE IF NOT EXISTS RecipeMalts (
    RecipeID INTEGER,
    MaltID INTEGER,
    Weight REAL,
    FOREIGN KEY (RecipeID) REFERENCES Recipes(ID),
    FOREIGN KEY (MaltID) REFERENCES Malts(ID)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS RecipeHops (
    RecipeID INTEGER,
    HopID INTEGER,
    Weight REAL,
    BoilingTime INTEGER,
    FOREIGN KEY (RecipeID) REFERENCES Recipes(ID),
    FOREIGN KEY (HopID) REFERENCES Hops(ID)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS RecipeStyles (
    RecipeID INTEGER,
    StyleID INTEGER,
    FOREIGN KEY (RecipeID) REFERENCES Recipes(ID),
    FOREIGN KEY (StyleID) REFERENCES Styles(ID)
)
''')


conn.commit()
conn.close()
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import sqlite3

# Define the custom InputError exception
class InputError(Exception):
    """Invalid input"""
    pass

# Function to fetch styles from the database
def fetch_styles():
    conn = sqlite3.connect('beer_recipes.db')
    cursor = conn.cursor()
    cursor.execute("SELECT ID, Name FROM Styles")
    styles = cursor.fetchall()
    conn.close()
    return [(style[1], style[0]) for style in styles]

# Widgets for Make a Recipe Tab
recipe_name_input = widgets.Text(
    description='Nazwa warki:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='300px')
)
date_input = widgets.DatePicker(
    description='Data tworzenia receptury:',
    disabled=False,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='300px')
)
style_dropdown = widgets.Dropdown(
    description='Styl:',
    options=fetch_styles(),
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='300px')
)
volume_input = widgets.FloatText(
    description='Objętość warki (L):',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='300px')
)
malt_water_factor_input = widgets.FloatText(
    description='Współczynnik woda/słód:',
    value=3.0,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='300px')
)
mashing_temperature_input = widgets.FloatText(
    description='Temperatura zacierania:',
    value=68.0,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='300px')
)

# Buttons
save_recipe_button = widgets.Button(description='Zapisz', layout=widgets.Layout(width='150px'))
clear_recipe_button = widgets.Button(description='Wyczyść', layout=widgets.Layout(width='150px'))
output_recipe = widgets.Output()

# Create a grid layout
grid_layout = widgets.GridspecLayout(2, 3, width='100%')

# Place widgets into the grid layout
grid_layout[0, 0] = recipe_name_input
grid_layout[0, 1] = date_input
grid_layout[0, 2] = style_dropdown
grid_layout[1, 0] = volume_input
grid_layout[1, 1] = malt_water_factor_input
grid_layout[1, 2] = mashing_temperature_input

# Title with increased font size
title = widgets.HTML(
    value="<h2 style='font-size:24px;'>Utwórz Recepturę</h2>",
    layout=widgets.Layout(margin='0 0 20px 0')
)

# Function to save a new recipe to the database
def save_recipe(*args):
    with output_recipe:
        clear_output(wait=True)
        try:
            name = recipe_name_input.value
            date = date_input.value
            volume = volume_input.value
            malt_water_factor = malt_water_factor_input.value
            mashing_temperature = mashing_temperature_input.value
            style_id = style_dropdown.value

            if not name or not date or not volume or not malt_water_factor or not mashing_temperature or not style_id:
                raise InputError("Uzupełnij wszystkie pola")

            if volume <= 0 or malt_water_factor <= 0 or mashing_temperature <= 0:
                raise ValueError("Wartości muszą być większe od zera")

            # Ensure the connection is opened here
            conn = sqlite3.connect('beer_recipes.db')
            cursor = conn.cursor()

            # Check if the recipe name already exists
            cursor.execute("SELECT COUNT(*) FROM Recipes WHERE Name = ?", (name,))
            if cursor.fetchone()[0] > 0:
                raise InputError("Podana nazwa już istnieje, użyj innej")

            # Insert the new recipe and get the recipe ID
            cursor.execute("INSERT INTO Recipes (Name, Date, Volume, MaltWaterFactor, MashingTemperature) VALUES (?, ?, ?, ?, ?)", (name, date, volume, malt_water_factor, mashing_temperature))
            recipe_id = cursor.lastrowid

            # Insert the style for the recipe
            cursor.execute("INSERT INTO RecipeStyles (RecipeID, StyleID) VALUES (?, ?)", (recipe_id, style_id))

            conn.commit()
            conn.close()  # Close the connection

            print(f"Receptura '{name}' została zapisana pomyślnie")

        except ValueError as e:
            print(f"Invalid input: {e}")

        except InputError as e:
            print(f"Input error: {e}")

        except sqlite3.Error as e:
            print(f"An error occurred: {e}")

        except Exception as e:
            print(f"An unexpected error occurred: {e}")

    # Update all dropdowns in other tabs
    update_recipe_dropdowns()

# Function to clear input fields
def clear_recipe(*args):
    recipe_name_input.value = ''
    date_input.value = None  # Reset date picker
    volume_input.value = 0.0  # Set to a default float value
    malt_water_factor_input.value = 0.0  # Set to a default float value
    mashing_temperature_input.value = 0.0  # Set to a default float value
    style_dropdown.value = None  # Reset the dropdown to None or any default option

save_recipe_button.on_click(save_recipe)
clear_recipe_button.on_click(clear_recipe)

# Arrange widgets vertically in a VBox
recipe_tab = widgets.VBox([
    title,
    grid_layout,
    widgets.HBox([save_recipe_button, clear_recipe_button]),
    output_recipe
])
```

## Zaciernie

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import sqlite3

# Function to fetch recipes from the database
def fetch_recipes():
    conn = sqlite3.connect('beer_recipes.db')
    cursor = conn.cursor()
    cursor.execute("SELECT ID, Name FROM Recipes")
    recipes = cursor.fetchall()
    conn.close()
    return [(recipe[1], recipe[0]) for recipe in recipes]

# Function to fetch malts from the database
def fetch_malts():
    conn = sqlite3.connect('beer_recipes.db')
    cursor = conn.cursor()
    cursor.execute("SELECT ID, Name FROM Malts")
    malts = cursor.fetchall()
    conn.close()
    return [(malt[1], malt[0]) for malt in malts]

recipe_malts = []

# Widgets for Mashing Tab
recipe_dropdown_mashing = widgets.Dropdown(
    description='Receptua', options=fetch_recipes(), style={'description_width': 'initial'}, layout=widgets.Layout(width='300px')
)
mashing_performance_input = widgets.FloatText(
    description='Wydajność zacierania', value = 80.0, style={'description_width': 'initial'}, layout=widgets.Layout(width='300px')
)
malt_dropdown = widgets.Dropdown(
    description='Słód:', options=fetch_malts(), style={'description_width': 'initial'}, layout=widgets.Layout(width='300px')
)
malt_weight_input = widgets.FloatText(
    description='Masa (kg)', style={'description_width': 'initial'}, layout=widgets.Layout(width='300px')
)

# Buttons
add_malt_button = widgets.Button(description='Dodaj słód', layout=widgets.Layout(width='150px'))
calculate_extraction_button = widgets.Button(description='Oblicz ekstrakcję', layout=widgets.Layout(width='200px'))
save_malt_button = widgets.Button(description='Zapisz', layout=widgets.Layout(width='150px'))
clear_malt_button = widgets.Button(description='Wyczyść', layout=widgets.Layout(width='150px'))

# Output areas
output_mashing = widgets.Output()
output_extraction = widgets.Output()

# Create a grid layout for the widgets
grid = widgets.GridspecLayout(2, 4, width='100%')

# Place widgets in the grid
grid[0, 0] = recipe_dropdown_mashing
grid[0, 1] = mashing_performance_input
grid[1, 0] = malt_dropdown
grid[1, 1] = malt_weight_input

# Title with increased font size
title = widgets.HTML(
    value="<h2 style='font-size:24px;'>Zacieranie</h2>",
    layout=widgets.Layout(margin='0 0 20px 0')
)

def calculate_extraction(*args):
    with output_extraction:
        clear_output(wait=True)
        if not recipe_malts:
            print("Brak słodów do obliczenia ekstrakcji")
            return None

        try:
            mashing_performance = mashing_performance_input.value
            if not mashing_performance or mashing_performance <= 0:
                print("Wartość wydajności jest niepoprawna")
                return None

            conn = sqlite3.connect('beer_recipes.db')
            cursor = conn.cursor()

            # Fetch the volume from the selected recipe
            recipe_id = recipe_dropdown_mashing.value
            if not recipe_id:
                print("Wybierz recepturę")
                return None

            cursor.execute("SELECT Volume FROM Recipes WHERE ID = ?", (recipe_id,))
            volume_row = cursor.fetchone()
            if volume_row is None:
                print("Nie znaleziono receptury")
                return None

            volume = volume_row[0]
            conn.close()

            total_extraction_g = sum(
                (malt['extraction'] * 10 * mashing_performance / 100) * malt['weight'] for malt in recipe_malts
            )
            extract_ml = total_extraction_g / 1.587
            water_volume_ml = volume * 1.1 * 1000 - extract_ml
            wort_weight_g = water_volume_ml + total_extraction_g
            extract_percent = total_extraction_g / wort_weight_g * 100

            print(f"Ekstrakcja wynosi: {extract_percent:.2f}%")
            return extract_percent
        except Exception as e:
            print("Podczas obliczania ekstrakcji wystąpił błąd", str(e))
            return None

def add_malt(*args):
    with output_mashing:
        clear_output(wait=True)
        malt_id = malt_dropdown.value
        weight = malt_weight_input.value

        if not malt_id:
            print("Wybierz słód")
            return
        if not weight or weight <= 0:
            print("Wprowadź dodatnią wartość masy słodu")
            return

        conn = sqlite3.connect('beer_recipes.db')
        cursor = conn.cursor()

        cursor.execute("SELECT Name, Extraction FROM Malts WHERE ID = ?", (malt_id,))
        malt_info = cursor.fetchone()
        conn.close()

        if malt_info:
            malt_name, extraction = malt_info
            recipe_malts.append({
                'malt_id': malt_id,
                'weight': weight,
                'extraction': extraction
            })
            print(f"Dodano słod: {malt_name}, w ilości: {weight} kg, jego ekstrakcja wynosi: {extraction}%")
        else:
            print("Nie znaleziono słodu w bazie danych")

def save_malt(*args):
    with output_mashing:
        clear_output()
        recipe_id = recipe_dropdown_mashing.value

        if not recipe_id:
            print("Aby zapisać recepturę, wybierz ją")
            return

        # Check for invalid or negative mashing performance
        mashing_performance = mashing_performance_input.value
        if not mashing_performance or mashing_performance <= 0:
            print("Wprowadź dodatnią wartość wydajności zacierania ")
            return

        # Check for invalid or negative malt weights
        for malt in recipe_malts:
            if malt['weight'] <= 0:
                print("Masa słodu pownna być dodatnią wartością")
                return

        if not recipe_malts:
            print("Nie wproawdzono słodu, zrób to przed zapisaniem")
            return

        extraction = calculate_extraction()
        if extraction is None:
            print("Nie obliczono ekstrakcji")
            return

        try:
            conn = sqlite3.connect('beer_recipes.db')
            cursor = conn.cursor()

            for malt in recipe_malts:
                cursor.execute("INSERT INTO RecipeMalts (RecipeID, MaltID, Weight) VALUES (?, ?, ?)",
                               (recipe_id, malt['malt_id'], malt['weight']))

            # Update extraction in Recipes table
            cursor.execute("UPDATE Recipes SET Extraction = ? WHERE ID = ?", (extraction, recipe_id))
            conn.commit()
            print("Dane słodów oraz ekstrakcja zostaly pomyślnie zapisane w bazie danych")
        except sqlite3.Error as e:
            print("Podczas zapisu, wystąpił błąd", e)
        finally:
            conn.close()

        # Clear the list of malts
        recipe_malts.clear()

def clear_malt(*args):
    mashing_performance_input.value = 0
    malt_weight_input.value = 0
    recipe_malts.clear()
    with output_mashing:
        clear_output()
    with output_extraction:
        clear_output()

add_malt_button.on_click(add_malt)
calculate_extraction_button.on_click(calculate_extraction)
save_malt_button.on_click(save_malt)
clear_malt_button.on_click(clear_malt)

# Arrange widgets vertically in a VBox
mashing_tab = widgets.VBox([
    title,
    grid,
    widgets.HBox([add_malt_button, calculate_extraction_button]),
    widgets.HBox([save_malt_button, clear_malt_button]),
    output_mashing,
    output_extraction
])
```

## Chmielenie

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import sqlite3

# Function to fetch recipes from the database
def fetch_recipes():
    conn = sqlite3.connect('beer_recipes.db')
    cursor = conn.cursor()
    cursor.execute("SELECT ID, Name FROM Recipes")
    recipes = cursor.fetchall()
    conn.close()
    return [(recipe[1], recipe[0]) for recipe in recipes]

# Function to fetch hops from the database
def fetch_hops():
    conn = sqlite3.connect('beer_recipes.db')
    cursor = conn.cursor()
    cursor.execute("SELECT ID, Name FROM Hops")
    hops = cursor.fetchall()
    conn.close()
    return [(hop[1], hop[0]) for hop in hops]

selected_hops = []

# ------------- Hops Tab ----------------
recipe_dropdown_hops = widgets.Dropdown(description='Receptura:', options=fetch_recipes())
hop_dropdown = widgets.Dropdown(description='Chmiel:', options=fetch_hops())
hop_weight_input = widgets.FloatText(description='Masa (g)')
boiling_time_input = widgets.IntText(description='Czas (min)')
add_hop_button = widgets.Button(description='Dodaj chmiel')
calculate_ibu_button = widgets.Button(description='Oblicz IBU')
save_hop_button = widgets.Button(description='Zapisz')
clear_hop_button = widgets.Button(description='Wyczyść')
output_hops = widgets.Output()
output_ibu = widgets.Output()
# Title with increased font size
title = widgets.HTML(
    value="<h2 style='font-size:24px;'>Chmielenie</h2>",
    layout=widgets.Layout(margin='0 0 20px 0')
)

# Function to calculate utilization based on boiling time
def calculate_utilization(time):
    max_utilization = 34
    table = [
        [9, 6],
        [19, 15],
        [29, 19],
        [44, 24],
        [59, 27],
        [74, 30],
    ]

    for boiling_time, utilization in table:
        if time < boiling_time:
            return utilization
    return max_utilization

# Function to calculate IBU
def calculate_ibu(*args):
    with output_ibu:
        clear_output()
        if not selected_hops:
            print("Nie dodano chmieli, nie można obliczyć IBU")
            return

        total_ibu = 0
        for hop in selected_hops:
            utilization = calculate_utilization(hop['time'])
            ibu = (hop['weight'] * hop['alpha_acid'] * utilization) / (volume_input.value * 10)
            total_ibu += ibu

        print(f"Obliczone IBU: {total_ibu:.2f}")
        return total_ibu

# Function to add hop
def add_hop(*args):
    with output_hops:
        clear_output(wait=True)
        try:
            hop_id = hop_dropdown.value
            weight = hop_weight_input.value
            time = boiling_time_input.value

            # Validate inputs
            if weight is None or weight <= 0:
                print("Masa musi być dodatnia wartością")
                return
            if time is None or time <= 0:
                print("Czas musi być dodatnia wartością")
                return

            conn = sqlite3.connect('beer_recipes.db')
            cursor = conn.cursor()

            cursor.execute("SELECT Name, AlphaAcids FROM Hops WHERE ID = ?", (hop_id,))
            hop_info = cursor.fetchone()
            conn.close()

            if hop_info is None:
                print("Nie znaleziono chmielu w bazie danych")
                return

            hop_name, alpha_acid = hop_info

            selected_hops.append({
                'hop_id': hop_id,
                'weight': weight,
                'time': time,
                'alpha_acid': alpha_acid
            })

            print(f"Dodano chmiel: {hop_name}, o masie: {weight} g, na czas: {time} min, zawartość alfakwasów: {alpha_acid}%")
        except Exception as e:
            print(f"error {e}")

# Function to save hop data and calculated IBU
def save_hop(*args):
    with output_hops:
        clear_output()
        try:
            recipe_id = recipe_dropdown_hops.value
            if not selected_hops:
                print("Nie dodano chmieli, dodaj zanim zapiszesz")
                return

            conn = sqlite3.connect('beer_recipes.db')
            cursor = conn.cursor()

            for hop in selected_hops:
                cursor.execute("INSERT INTO RecipeHops (RecipeID, HopID, Weight, BoilingTime) VALUES (?, ?, ?, ?)",
                               (recipe_id, hop['hop_id'], hop['weight'], hop['time']))

            # Calculate IBU and update it in Recipes table
            ibu = calculate_ibu()
            if ibu is not None:
                cursor.execute("UPDATE Recipes SET IBU = ? WHERE ID = ?", (ibu, recipe_id))
                conn.commit()
                print("Dane chmieli i IBU zostały pomyślnie zapisane w bazie danych")
            else:
                print("IBU nie zostało obliczone, brak inforamcji o chmielach")

            conn.close()
            selected_hops.clear()

        except Exception as e:
            print(f"An error occurred: {e}")

# Function to clear hop fields
def clear_hop(*args):
    hop_weight_input.value = 0
    boiling_time_input.value = 0
    selected_hops.clear()
    with output_hops:
        clear_output()
    with output_ibu:
        clear_output()

add_hop_button.on_click(add_hop)
calculate_ibu_button.on_click(calculate_ibu)
save_hop_button.on_click(save_hop)
clear_hop_button.on_click(clear_hop)

hops_tab = widgets.VBox([
    title,
    recipe_dropdown_hops,
    hop_dropdown,
    hop_weight_input,
    boiling_time_input,
    add_hop_button,
    calculate_ibu_button,
    save_hop_button,
    clear_hop_button,
    output_hops,
    output_ibu
])


```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, Javascript
import time
import sqlite3

# Function to play a sound
def play_sound():
    display(Javascript('''
        var audio = new Audio('https://www.soundjay.com/buttons/sounds/beep-05.mp3');
        audio.play().catch(function(error) {
            alert("Koniec czasu");
        });
    '''))

# Timer function with minutes and seconds display
def start_timer(duration, hop_name, current_hop_name):
    for i in range(duration, 0, -1):
        minutes, seconds = divmod(i, 60)
        time_display.value = f'{current_hop_name} teraz sie gotuje. Pozostało: {minutes} minut {seconds} sekund'
        time.sleep(1)
    play_sound()
    time_display.value = f'Dodaj {hop_name} teraz!'

# Fetch recipes, malts, and hops functions
def fetch_recipes():
    conn = sqlite3.connect('beer_recipes.db')
    cursor = conn.cursor()
    cursor.execute("SELECT ID, Name FROM Recipes")
    recipes = cursor.fetchall()
    conn.close()
    return [(recipe[1], recipe[0]) for recipe in recipes]

def fetch_malts_by_recipe(recipe_id):
    conn = sqlite3.connect('beer_recipes.db')
    cursor = conn.cursor()
    cursor.execute("""
        SELECT m.Name, rm.Weight, m.Extraction
        FROM RecipeMalts rm
        JOIN Malts m ON rm.MaltID = m.ID
        WHERE rm.RecipeID = ?
    """, (recipe_id,))
    malts = cursor.fetchall()
    conn.close()
    return malts

def fetch_hops_by_recipe(recipe_id):
    conn = sqlite3.connect('beer_recipes.db')
    cursor = conn.cursor()
    cursor.execute("""
        SELECT h.Name, rh.Weight, rh.BoilingTime, h.AlphaAcids
        FROM RecipeHops rh
        JOIN Hops h ON rh.HopID = h.ID
        WHERE rh.RecipeID = ?
    """, (recipe_id,))
    hops = cursor.fetchall()
    conn.close()
    return hops

# Brew it! Tab
recipe_dropdown_brew = widgets.Dropdown(description='Receptura:', options=fetch_recipes())
brew_button = widgets.Button(description="Uwarz to!")
output_brew = widgets.Output()
time_display = widgets.Label(value='')
start_button = widgets.Button(description='Uruchom minutnik')
title = widgets.HTML(
    value="<h2 style='font-size:24px;'>Uwarz to!</h2>",
    layout=widgets.Layout(margin='0 0 20px 0')
)

# Start button click handler
def start_button_click_handler(_):
    start_next_hop_timer(0)

# Function to start the next hop timer
def start_next_hop_timer(index):
    if index < len(hops_sorted) - 1:
        hop_name, hop_weight, boiling_time, alpha_acid = hops_sorted[index]
        next_hop_time = hops_sorted[index + 1][2]
        wait_time = boiling_time - next_hop_time
        current_hop_name = hop_name
        start_timer(wait_time * 60, hops_sorted[index + 1][0], current_hop_name)
        start_next_hop_timer(index + 1)
    else:
        # Last hop
        hop_name, hop_weight, boiling_time, alpha_acid = hops_sorted[index]
        current_hop_name = hop_name
        start_timer(boiling_time * 60, hop_name, current_hop_name)
        time_display.value = 'Gotowanie zakończone, czas schłodzić brzeczkę'

def display_brew_guide(*args):
    global hops_sorted
    with output_brew:
        clear_output(wait=True)

        # Get selected recipe ID
        recipe_id = recipe_dropdown_brew.value
        if not recipe_id:
            print("Wybierz recepturę")
            return

        # Fetch recipe details
        conn = sqlite3.connect('beer_recipes.db')
        cursor = conn.cursor()
        cursor.execute("""
            SELECT Name, Date, Volume, Extraction, IBU, MashingTemperature, MaltWaterFactor
            FROM Recipes
            WHERE ID = ?
        """, (recipe_id,))
        recipe_info = cursor.fetchone()
        conn.close()

        if not recipe_info:
            print("Brak informacji o recepturze.")
            return

        (name, date, volume, extraction, ibu, mashing_temperature, malt_water_factor) = recipe_info

        # Display recipe information
        print(f"Warzysz teraz: {name}")
        print(f"Recptura została zapisana: {date}")
        print(f"Objętość twojej warki: {volume} L")
        print(f"Współczynnik woda/słód: {malt_water_factor}")
        print(f"Temperatura zacierania: {mashing_temperature} °C")
        print(f"Ekstrakcja brzeczki wyniesie: {extraction} Blg")

        # Mashing Process
        print("\n----------- Zacieranie -----------")
        malts = fetch_malts_by_recipe(recipe_id)
        total_malt_weight = sum(malt[1] for malt in malts)
        for malt_name, malt_weight, extraction in malts:
            print(f"Słód: {malt_name}, Masa: {malt_weight} kg, Ekstrakcja: {extraction}%")

        # Water calculations
        water_for_mashing = total_malt_weight * malt_water_factor
        malt_temperature = 20  # Malt temperature is assumed to be 20°C
        water_for_mashing_temperature = mashing_temperature + \
            ((0.5 * (mashing_temperature - malt_temperature) * total_malt_weight) / water_for_mashing)

        malt_losses = total_malt_weight
        boiling_losses = volume * 0.15
        hop_yeast_losses = volume * 0.1
        rest_water = volume - water_for_mashing
        water_for_lautering = malt_losses + boiling_losses + hop_yeast_losses + rest_water

        print(f"\nWoda potrzebna do zacierania wynosi: {water_for_mashing:.2f} L")
        print(f"Temperatura do której powinieneś podgrzać wodę przed wsypaniem słodu wynosi: {water_for_mashing_temperature:.2f} °C")
        print(f"Woda potrzebna do wysładzania: {water_for_lautering:.2f} L")
        print("Zacieraj przez około 60 minut, następnie przejdź do filtracji i wysładzania")
        # Wort Boiling Process
        print("\n----------- Chmielenie -----------")
        hops = fetch_hops_by_recipe(recipe_id)
        hops_sorted = sorted(hops, key=lambda x: x[2], reverse=True)  # Sort by boiling time, descending
        print(f"\n IBU twojej warki wynosi {ibu}")

        # Display hops information
        for hop_name, hop_weight, boiling_time, alpha_acid in hops_sorted:
            print(f"Chmiel: {hop_name}, Masa: {hop_weight} g, Czas gotowania: {boiling_time} min, Alfa kwasy: {alpha_acid}%")

        # Show start button
        print("\nGdy brzeczka się zagotuje, przygotuj pierwszy (o najdłuższym czasie gotowania) chmiel, dodając go wystartuj minutnik")
        print("Po schłodzeniu brzeczki do 30 °C dodaj drożdże odstaw brzeczkę do fermentacji")
        print("Po procesie fermentacji, zmierz Blg i przejdź do zakładki ABV")
        display(start_button)


start_button.on_click(start_button_click_handler)

brew_tab = widgets.VBox([
    title,
    recipe_dropdown_brew,
    brew_button,
    time_display,
    output_brew
])



brew_button.on_click(display_brew_guide)
```

## ABV

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import sqlite3

# Fetch recipes from the database with their extraction
def fetch_recipes():
    conn = sqlite3.connect('beer_recipes.db')
    cursor = conn.cursor()
    cursor.execute("SELECT ID, Name FROM Recipes")
    recipes = cursor.fetchall()
    conn.close()
    return [(recipe[1], recipe[0]) for recipe in recipes]

# Widgets for the new tab
recipe_dropdown_abv = widgets.Dropdown(description='Receptura:', options=fetch_recipes())
extraction_after_fermentation_input = widgets.FloatText(
    description='Ekstrakcja po fermentacji:', style={'description_width': 'initial'}, layout=widgets.Layout(width='300px')
)
abv_output = widgets.Label(value='ABV: N/A')
calculate_abv_button = widgets.Button(description='Oblicz ABV')
save_button = widgets.Button(description='Zapisz')
output_abv = widgets.Output()
title = widgets.HTML(
    value="<h2 style='font-size:24px;'>Oblicz ABV</h2>",
    layout=widgets.Layout(margin='0 0 20px 0')
)

# Function to calculate ABV
def calculate_abv(*args):
    recipe_id = recipe_dropdown_abv.value
    if not recipe_id:
        abv_output.value = 'Wybierz recepturę'
        return

    extraction_after_fermentation = extraction_after_fermentation_input.value

    # Fetch extraction from the selected recipe
    conn = sqlite3.connect('beer_recipes.db')
    cursor = conn.cursor()
    cursor.execute("SELECT Extraction FROM Recipes WHERE ID = ?", (recipe_id,))
    extraction = cursor.fetchone()[0]
    conn.close()

    # Calculate ABV
    abv = (extraction - extraction_after_fermentation) / 1.9047
    abv_output.value = f'ABV: {abv:.2f}%'

# Function to save ABV and Extraction After Fermentation to the database
def save_abv(*args):
    recipe_id = recipe_dropdown_abv.value
    if not recipe_id:
        abv_output.value = 'Wybierz recepturę'
        return

    extraction_after_fermentation = extraction_after_fermentation_input.value

    # Fetch extraction from the selected recipe
    conn = sqlite3.connect('beer_recipes.db')
    cursor = conn.cursor()
    cursor.execute("SELECT Extraction FROM Recipes WHERE ID = ?", (recipe_id,))
    extraction = cursor.fetchone()[0]

    # Calculate ABV
    abv = (extraction - extraction_after_fermentation) / 1.9047

    # Save Extraction After Fermentation and ABV to the database
    cursor.execute("""
        UPDATE Recipes
        SET ExtractionAfterFermentation = ?, ABV = ?
        WHERE ID = ?
    """, (extraction_after_fermentation, abv, recipe_id))
    conn.commit()
    conn.close()

    abv_output.value = 'Dane zostały pomyślnie zapisane w bazie'

calculate_abv_button.on_click(calculate_abv)
save_button.on_click(save_abv)

# ABV Calculation and Save Tab
abv_tab = widgets.VBox([
    title,
    recipe_dropdown_abv,
    extraction_after_fermentation_input,
    calculate_abv_button,
    abv_output,
    save_button,
    output_abv
])

```
